```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
 
# To read image from disk, we use
# cv2.imread function, in below method,
img = cv2.imread("imagem-simulacao.jpg")
blurred_image = cv2.GaussianBlur(img, (75, 75), 0)  # You can adjust the kernel size (e.g., (5, 5)) and the standard deviation (0) as needed
plt.imshow(img)
```

```python
height, width, _ = blurred_image.shape
crop_width = 580  # Specify the width of the cropped region
crop_height = 580  # Specify the height of the cropped region

# Calculate the coordinates for the top-left corner of the cropped region
x = (width - crop_width) // 2
y = (height - crop_height) // 2

# Crop the image based on the calculated coordinates
cropped_image = img[y:y + crop_height, x:x + crop_width]
plt.imshow(cropped_image)
```

```python
hsv_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2HSV)
# Create a Matplotlib figure and axes
fig, ax = plt.subplots()

# Display the HSV image with a colorbar
cax = ax.imshow(hsv_image[:, :, 0], cmap='hsv', vmin=0, vmax=179)  # Display the H channel (Hue)
plt.colorbar(cax, ax=ax, orientation='vertical', label='Hue (0-179)')  # Add a colorbar

# Set axis labels and title (you can customize these as needed)
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Imagem HSV')

# Show the plot
plt.show()
```

```python
# Define the lower and upper hue thresholds
lower_hue = 100
upper_hue = 140

# Create a mask based on the hue channel
hue_channel = hsv_image[:, :, 0]
mask_hue = cv2.inRange(hue_channel, lower_hue, upper_hue)

# Define a kernel for morphological operations (adjust the size as needed)
kernel = np.ones((5, 5), np.uint8)  # 5x5 kernel

# Apply dilation to include neighboring pixels
dilated_mask = cv2.dilate(mask_hue, kernel, iterations=1)

# Create a Matplotlib figure and axes
fig, ax = plt.subplots()

# Display the masked HSV image
masked_image = cv2.bitwise_and(hsv_image, hsv_image, mask=dilated_mask)
masked_rgb_image = cv2.cvtColor(masked_image, cv2.COLOR_HSV2RGB)
cax = ax.imshow(masked_rgb_image)

# Set axis labels and title (you can customize these as needed)
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Masked HSV Image')

# Show the plot
plt.show()
```

```python
# Apply the mask to the original image using NumPy indexing
masked_original_image = cropped_image.copy()
masked_original_image[dilated_mask == 0] = 0

# Create a Matplotlib figure and axes
fig, ax = plt.subplots()

# Display the masked original image
cax = ax.imshow(cv2.cvtColor(masked_original_image, cv2.COLOR_BGR2RGB))

# Set axis labels and title (you can customize these as needed)
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Masked Original Image')

# Show the plot
plt.show()
```

```python

image = cv2.imread("imagem-simulacao.jpg", cv2.IMREAD_GRAYSCALE)
image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)

# Apply a binary threshold to create an initial mask
_, initial_mask = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY)

# Define a kernel for morphological operations (adjust the size as needed)
kernel = np.ones((3, 3), np.uint8)  # 3x3 kernel

# Apply dilation to include neighboring pixels
dilated_mask = cv2.dilate(initial_mask, kernel, iterations=1)

# Create a Matplotlib figure and axes
fig, ax = plt.subplots()

# Display the masked image with the original grayscale colormap
masked_image = cv2.bitwise_and(image, image, mask=dilated_mask)
cax = ax.imshow(masked_image, cmap='gray', vmin=0, vmax=255)  # Use 'gray' colormap for grayscale
plt.colorbar(cax, ax=ax, orientation='vertical', label='Intensity')  # Add a colorbar

# Set axis labels and title (you can customize these as needed)
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Masked Grayscale Image with Colorbar')

# Show the plot
plt.show()
```

```python
import cv2
import numpy as np

# Create a VideoCapture object to read the video
cap = cv2.VideoCapture('simulacao.mp4')  # Replace 'your_input_video.mp4' with your video file path

# Check if the video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Create a VideoWriter object to save the processed video
frame_width = int(cap.get(3))  # Width of the frames
frame_height = int(cap.get(4))  # Height of the frames
out = cv2.VideoWriter('output_video.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 30, (frame_width, frame_height))

# Define the lower and upper hue thresholds
lower_hue = 100
upper_hue = 140

# Define a kernel for morphological operations (adjust the size as needed)
kernel = np.ones((5, 5), np.uint8)  # 5x5 kernel

while True:
    # Read a frame from the video
    ret, frame = cap.read()

    # Break the loop if we've reached the end of the video
    if not ret:
        break

    # Convert the frame to HSV
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Create a mask based on the hue channel
    hue_channel = hsv_frame[:, :, 0]
    mask_hue = cv2.inRange(hue_channel, lower_hue, upper_hue)

    # Apply dilation to include neighboring pixels
    dilated_mask = cv2.dilate(mask_hue, kernel, iterations=1)

    # Apply the mask to the original frame using NumPy indexing
    masked_frame = frame.copy()
    masked_frame[dilated_mask == 0] = 0

    # Write the processed frame to the output video
    out.write(masked_frame)

    # Display the processed frame (optional)
    cv2.imshow('Processed Video', masked_frame)

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video objects
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb

# Create a VideoCapture object to read the video
cap = cv2.VideoCapture('simulacao.mp4')  # Replace 'your_input_video.mp4' with your video file path

# Check if the video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Create a VideoWriter object to save the processed video
frame_width = int(cap.get(3))  # Width of the frames
frame_height = int(cap.get(4))  # Height of the frames
out = cv2.VideoWriter('simulacao.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 30, (frame_width, frame_height))

# Create a Matplotlib figure and axes for displaying the video with a colorbar
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Define the lower and upper hue thresholds
lower_hue = 100
upper_hue = 140

# Define a kernel for morphological operations (adjust the size as needed)
kernel = np.ones((5, 5), np.uint8)  # 5x5 kernel

while True:
    # Read a frame from the video
    ret, frame = cap.read()

    # Break the loop if we've reached the end of the video
    if not ret:
        break

    # Convert the frame to HSV
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Create a mask based on the hue channel
    hue_channel = hsv_frame[:, :, 0]
    mask_hue = cv2.inRange(hue_channel, lower_hue, upper_hue)

    # Apply dilation to include neighboring pixels
    dilated_mask = cv2.dilate(mask_hue, kernel, iterations=1)

    # Find contours in the dilated mask
    contours, _ = cv2.findContours(dilated_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the frame
    frame_with_contours = frame.copy()
    cv2.drawContours(frame_with_contours, contours, -1, (0, 255, 0), 2)  # Green contours

    # Convert the HSV frame to RGB for display
    hsv_rgb_frame = cv2.cvtColor(hsv_frame, cv2.COLOR_HSV2RGB)

    # Display the HSV frame with a colorbar
    ax[0].imshow(hsv_rgb_frame)
    ax[0].set_xlabel('X-Axis')
    ax[0].set_ylabel('Y-Axis')
    ax[0].set_title('HSV Video')

    # Display the processed frame with contours
    ax[1].imshow(cv2.cvtColor(frame_with_contours, cv2.COLOR_BGR2RGB))
    ax[1].set_xlabel('X-Axis')
    ax[1].set_ylabel('Y-Axis')
    ax[1].set_title('Processed Video with Contours')

    # Update the colorbar
    cax = ax[0].inset_axes([0.1, 0.1, 0.03, 0.6])
    cbar = plt.colorbar(hsv_to_rgb(np.array([[[lower_hue / 179, 1, 1]]])), cax=cax, orientation='vertical')
    cbar.set_label('Hue')

    # Update the Matplotlib figure
    plt.pause(0.01)
    plt.draw()

    # Write the processed frame with contours to the output video
    out.write(frame_with_contours)

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video objects
cap.release()
out.release()
cv2.destroyAllWindows()
```
